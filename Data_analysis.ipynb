```python
# importing neccessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram 

data_f = np.loadtxt('flows.txt') # loading data flows.txt

plt.subplot(1, 4, 2)
sns.histplot(data_f, bins=10, kde=True)
plt.title('Hist')
plt.show()
```

```python
# Scatter plot

plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.scatter(range(len(data_f)), data_f, s=5)
plt.title('Scatter plot')
plt.show()
```

```python
# Box plot 

plt.subplot(1, 4, 3)
sns.boxplot(data_f, orient='vertical')
plt.title('Boxplot')
plt.show()
```

```python
# ECDF

plt.subplot(1, 4, 4)
sns.ecdfplot(data_f)
plt.title('Empirical CDF')
plt.show()
```

Task 2(2)

```python
mean = np.mean(data_f)
median = np.median(data_f)
max_value = np.max(data_f)
min_value = np.min(data_f)

print("Variable 1 (flows.txt data) Summary:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: N/A")
print(f"Max: {max_value:.2f}")
print(f"Min: {min_value:.2f}")
```

Task 2(3)

```python
log_data_f = np.log(data_f)

#histogram with log data

plt.subplot(1, 4, 2)
sns.histplot(log_data_f, bins=20, kde=True)
plt.title('Log Histogram')
plt.show()
```

```python
# scatter plot with log data

plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.scatter(range(len(log_data_f)), log_data_f, s=5)
plt.title('Log Scatterplot')
plt.show()
```

```python
# Box plot with log data

plt.subplot(1, 4, 3)
sns.boxplot(log_data_f, orient='vertical')
plt.title('Log Boxplot')
plt.show()
```

```python
# ECDF plot with log data

plt.subplot(1, 4, 4)
sns.ecdfplot(log_data_f)
plt.title('Log Empirical CDF')

plt.show()
```
